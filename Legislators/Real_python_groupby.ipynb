```python
import pandas as pd
```

```python
dtypes = {'first_name': 'category',
         'gender': 'category',
         'type': 'category',
         'state': 'category',
         'party': 'category',}
```

```python
df = pd.read_csv('legislators-historical.csv', dtype=dtypes,
                usecols=list(dtypes) + ['birthday', 'last_name'],
                parse_dates=['birthday'])
```
